```python
import pandas as pd
import numpy as np
from statsmodels.stats.multitest import multipletests
```

```python
def color_values(val):
    color = 'crimson' if val >= 0.05 else 'royalblue'
    return f'color: {color}'

def color_values_ref(val):
    color = 'crimson' if val >= 0.05 else 'darkorange'
    return f'color: {color}'
```

## Load Data

```python
df = pd.read_excel('../Data/p-values-non-correction.xlsx')
```

```python
df["FDR"] = multipletests(df["p-value"], method='fdr_bh')[1]


df = df.round(3)
df = (
    df.style
    .applymap(color_values, subset=["p-value", "FDR"])
    .applymap(color_values_ref, subset=["p-value"])
)
```
