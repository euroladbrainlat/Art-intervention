## Load Data

```python
import pickle

with open("../Data/child_self_expression.pkl", "rb") as f:
    text_concat = pickle.load(f)
```

```python
nltk.download('stopwords')
spanish_stopwords = stopwords.words('english')

wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=set(spanish_stopwords)).generate(text_concat)
plt.figure(figsize=(5, 3))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()
```

```python
vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words=spanish_stopwords)
X = vectorizer.fit_transform([text_concat])
    
bigramas = vectorizer.get_feature_names_out()
frecuencias = X.toarray().sum(axis=0)
        

bigramas_frecuencias = dict(zip(bigramas, frecuencias))

        
wordcloud = WordCloud(width=1200, height=800, background_color='white', colormap='Dark2').generate_from_frequencies(bigramas_frecuencias)
plt.figure(figsize=(5, 3))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()
```
